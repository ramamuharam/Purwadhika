### 1. Toy Datasets: Boston house prices dataset

506 data with 13 features

```python
from sklearn.datasets import load_boston
```

```python
dataBoston = load_boston()
```

```python
# dir(dataBoston) # ['DESCR', 'data', 'feature_names', 'filename', 'target']
print(dataBoston['data'][0]) # value dari features (X) = 506 data, 13 feature
print(dataBoston['feature_names'])
print(dataBoston['target'][0])
```

```python
dfBoston = pd.DataFrame(
    dataBoston['data'],
    columns = dataBoston['feature_names']
)
dfBoston['PRICE'] = dataBoston['target']
dfBoston
```

```python
# Pearson Correlation
dfBoston.corr()
```

```python
# import seaborn as sb
# sb.pairplot(dfBoston)
# plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# training: all data, 5 features: LSTAT, PTRATIO, TAX, RM, INDUS
model.fit(dfBoston[['LSTAT', 'PTRATIO', 'TAX', 'RM', 'INDUS']], dfBoston['PRICE'])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
dfBoston['PRICEpred'] = model.predict(dfBoston[['LSTAT', 'PTRATIO', 'TAX', 'RM', 'INDUS']])
dfBoston[dfBoston['PRICEpred'] < 0] = 0
dfBoston
```

```python
# Evaluation Metrics using Sklearn
from sklearn.metrics import (
    max_error,
    mean_absolute_error, 
    mean_squared_error, 
    mean_squared_log_error,
    median_absolute_error,
    r2_score
)
```

```python
print(f"Max Error = {max_error(dfBoston['PRICE'], dfBoston['PRICEpred'])}")
print(f"MAE = {mean_absolute_error(dfBoston['PRICE'], dfBoston['PRICEpred'])}")
print(f"MSE = {mean_squared_error(dfBoston['PRICE'], dfBoston['PRICEpred'])}")
print(f"RMSE = {np.sqrt(mean_squared_error(dfBoston['PRICE'], dfBoston['PRICEpred']))}")
print(f"MSLE = {mean_squared_log_error(dfBoston['PRICE'], dfBoston['PRICEpred'])}")
print(f"RMSLE = {np.sqrt(mean_squared_log_error(dfBoston['PRICE'], dfBoston['PRICEpred']))}")
print(f"MedAE = {median_absolute_error(dfBoston['PRICE'], dfBoston['PRICEpred'])}")
print(f"R2 score = {r2_score(dfBoston['PRICE'], dfBoston['PRICEpred'])}")
```
