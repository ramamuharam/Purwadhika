### E. Binarizer

- Merubah data menjadi data biner (value 0 atau 1) aka boolean dengan menggunakan threshold tertentu. Misal diambil threshold = 20, maka data dengan value <= 20 akan berubah jadi 0 & data value > 20 akan menjadi 1.

````python
df = pd.DataFrame({
    'TB': [1.72, 1.81, 1.93, 1.67, 1.85, 1.66, 1.59, 1.76, 1.88, 1.78],
    'BB': [76, 65, 87, 55, 60, 78, 98, 77, 74, 64]
})
df.head(2)
````

<hr>

### E1. Manual Calculation

Set threshold TB = 1.80 & threshold BB = 80

````python
df['bTB'] = df['TB'].apply(lambda x: 0 if x <= 1.8 else 1)
df['bBB'] = df['BB'].apply(lambda x: 0 if x <= 80 else 1)
df
````

<hr>

### E2. Sklearn ```Binarizer```

Set threshold TB = 1.80 & threshold BB = 80

````python
from sklearn.preprocessing import Binarizer
````

````python
transformer = Binarizer(threshold=1.8)
transformer.fit(df[['TB']])
df['bTB2'] = transformer.transform(df[['TB']]).astype('int32')
````

````python
transformer = Binarizer(threshold=80)
transformer.fit(df[['BB']])
df['bBB2'] = transformer.transform(df[['BB']])
df
````
